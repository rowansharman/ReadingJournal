Write a method belonging to `Deck` called `deal_hands` that takes two parameters, the number of hands and the number of cards per hand, and that creates new `Hand` objects, deals the appropriate number of cards per hand, and returns a list of `Hand` objects.

```python
"""This module contains a code example related to

Think Python, 2nd Edition
by Allen Downey
http://thinkpython2.com

Copyright 2015 Allen Downey

License: http://creativecommons.org/licenses/by/4.0/
"""

from __future__ import print_function, division

import random


class Card:
    """Represents a standard playing card.
    
    Attributes:
      suit: integer 0-3
      rank: integer 1-13
    """

    suit_names = ["Clubs", "Diamonds", "Hearts", "Spades"]
    rank_names = [None, "Ace", "2", "3", "4", "5", "6", "7", 
              "8", "9", "10", "Jack", "Queen", "King", "Ace"]

    def __init__(self, suit=0, rank=2):
        self.suit = suit
        self.rank = rank

    def __str__(self):
        """Returns a human-readable string representation."""
        return '%s of %s' % (self.rank_names[self.rank],
                             self.suit_names[self.suit])

    def __eq__(self, other):
        """Checks whether self and other have the same rank and suit.

        returns: boolean
        """
        return self.suit == other.suit and self.rank == other.rank

    def __lt__(self, other):
        """Compares this card to other, first by suit, then rank.

        returns: boolean
        """
        t1 = self.rank, self.suit
        t2 = other.rank, other.suit
        return t1 < t2


class Deck:
    """Represents a deck of cards.

    Attributes:
      cards: list of Card objects.
    """
    
    def __init__(self):
        """Initializes the Deck with 52 cards.
        """
        self.cards = []
        for suit in range(4):
            for rank in range(1, 14):
                card = Card(suit, rank)
                self.cards.append(card)

    def __str__(self):
        """Returns a string representation of the deck.
        """
        res = []
        for card in self.cards:
            res.append(str(card))
        return '\n'.join(res)

    def add_card(self, card):
        """Adds a card to the deck.

        card: Card
        """
        self.cards.append(card)

    def remove_card(self, card):
        """Removes a card from the deck or raises exception if it is not there.
        
        card: Card
        """
        self.cards.remove(card)

    def pop_card(self, i=-1):
        """Removes and returns a card from the deck.

        i: index of the card to pop; by default, pops the last card.
        """
        return self.cards.pop(i)

    def shuffle(self):
        """Shuffles the cards in this deck."""
        random.shuffle(self.cards)

    def sort(self):
        """Sorts the cards in ascending order."""
        self.cards.sort()

    def move_cards(self, hand, num):
        """Moves the given number of cards from the deck into the Hand.

        hand: destination Hand object
        num: integer number of cards to move
        """
        for i in range(num):
            hand.add_card(self.pop_card())
            
    def deal_hands(self, numHands, cards):
        hands = []
        for i in range(numHands):
            hand = Hand()
            self.move_cards(hand, cards)
            hand.sort()
            hands.append(hand)
        return hands
    
    
    def deal_poker_hands(self, numHands, cards):
        hands = []
        for i in range(numHands):
            hand = PokerHand()
            self.move_cards(hand, cards)
            hand.sort()
            hands.append(hand)
        return hands


class Hand(Deck):
    """Represents a hand of playing cards."""
    
    def __init__(self, label=''):
        self.cards = []
        self.label = label


def find_defining_class(obj, method_name):
    """Finds and returns the class object that will provide 
    the definition of method_name (as a string) if it is
    invoked on obj.

    obj: any python object
    method_name: string method name
    """
    for ty in type(obj).mro():
        if method_name in ty.__dict__:
            return ty
    return None


if __name__ == '__main__':
    deck = Deck()
    deck.shuffle()
    # deck.sort()

    hands = deck.deal_hands(3,5)
    for i in range(len(hands)):
        hands[i].label = str(i)
        print("\nPlayer %s's hand:" %(hands[i].label))
        print(hands[i])
```

The following are the possible hands in poker, in increasing order of value (and decreasing order of probability):

 1. **pair:** two cards with the same rank 
 2. **two pair:** two pairs of cards with the same rank 
 3. **three of a kind:** three cards with the same rank 
 4. **straight:** five cards with ranks in sequence (aces can be high or low, so Ace-2-3-4-5 is a straight and so is 10-Jack-Queen-King-Ace, but Queen-King-Ace-2-3 is not.) 
 5. **flush:** five cards with the same suit 
 6. **full house:** three cards with one rank, two cards with another 
 7. **four of a kind:** four cards with the same rank 
 8. **straight flush:** five cards in sequence (as defined above) and with the same suit

```python
class PokerHand(Hand):
    """Represents a poker hand."""

    def suit_hist(self):
        """Builds a histogram of the suits that appear in the hand.

        Stores the result in attribute suits.
        """
        self.suits = {}
        for card in self.cards:
            self.suits[card.suit] = self.suits.get(card.suit, 0) + 1
    
    def rank_hist(self):
        """Builds a histogram of the ranks that appear in the hand.

        Stores the result in attribute ranks.
        """
        self.ranks = {}
        for card in self.cards:
            self.ranks[card.rank] = self.ranks.get(card.rank, 0) + 1

    def has_pair(self):
        self.rank_hist()
        for val in self.ranks.values():
            if val >= 2:
                return True
        return False
    
    def has_two_pair(self):
        numPairs = 0
        self.rank_hist()
        for val in self.ranks.values():
            if val >= 2:
                numPairs += 1
        if numPairs >= 2:
            return True
        return False
    
    def has_three_kind(self):
        self.rank_hist()
        for val in self.ranks.values():
            if val >= 3:
                return True
        return False
    
    def has_straight(self):
        self.rank_hist()
        count = 0
        ranks = []
        for item in self.ranks:
            ranks.append(item)
        if 1 in ranks:
            ranks.append(14)
        ranks.sort()
        for i in range(1, len(ranks)):
            if ranks[i] == ranks[i-1] + 1:
                count += 1
                if count >= 4:
                    return True
            else:
                count = 0
        
        return False
    
    def has_flush(self):
        """Returns True if the hand has a flush, False otherwise.
      
        Note that this works correctly for hands with more than 5 cards.
        """
        self.suit_hist()
        for val in self.suits.values():
            if val >= 5:
                return True
        return False
    
    def has_full_house(self):
        if self.has_two_pair() and self.has_three_kind():
            return True
        return False
    
    def has_four_kind(self):
        self.rank_hist()
        for val in self.ranks.values():
            if val >= 4:
                return True
        return False
    
    def has_straight_flush(self):
        tempHand = self
        tempHand.sort()
        
        if tempHand.cards[0].rank == 1:
            aceHigh = Card(tempHand.cards[0].suit, 14)
            tempHand.add_card(aceHigh)
        
        ranks = []
        for item in tempHand.cards:
            ranks.append(item.rank)
            
        straights = [3]  # indicies of the last card in a straight of 5
        running = True
        runCount = 0
        
        while running:
            count = 0
            runCount += 1
            for i in range(straights[-1]-2, len(ranks)):  # start at the index one more than the
                if ranks[i] == ranks[i-1] + 1:            # beginning of the last straight found
                    count += 1
                    if count == 4:  # 4 matches = a sequence of 5
                        straights.append(i)
                else:
                    count = 0

            
            if len(straights) == 1:  # if there are no straights, there are no straight flushes
                return False

            if runCount == len(straights):
                running = False
        
        straights.remove(3)
        print(straights)
        for straightInd in straights:
            straightCards = []
            for i in range(0,4):
                straightCards.append(tempHand.cards[straightInd - i])
                straightHand = PokerHand(straightCards)
                print(straightHand)
                if straightHand.has_flush():  # check to see if the straight of 5 is also a flush
                    return True
        return False
    
    def classify(self):
        if self.has_straight_flush():
            return 0
        if self.has_four_kind():
            return 1
        if self.has_full_house():
            return 2
        if self.has_flush():
            return 3
        if self.has_straight():
            return 4
        if self.has_three_kind():
            return 5
        if self.has_two_pair():
            return 6
        if self.has_pair():
            return 7
        return 8
    


if __name__ == '__main__':
    
    totals = [0,0,0,0,0,0,0,0,0]
    decks = int(1000)
    
    for i in range(decks):
        # make a deck
        deck = Deck()
        deck.shuffle()

        # deal the cards and classify the hands

        hands = deck.deal_poker_hands(7, 7)
        for i in range(len(hands)):
#             hands[i].label = str(i)
#             print("\nPlayer %s's hand:" %(hands[i].label))
#             print(hands[i])
#             print(hands[i].has_straight_flush())
            totals[hands[i].classify()] += 1
    
print('In total of %i hands:' % (decks*7))
print('\tStraight flush   ', totals[0])
print('\tFour of a kind   ', totals[1])
print('\tFull house:      ', totals[2])
print('\tFlush            ', totals[3])
print('\tStraight:        ', totals[4])
print('\tThree of a kind: ', totals[5])
print('\tTwo pair:        ', totals[6])
print('\tOne pair:        ', totals[7])
print('\tNothing:         ', totals[8])
            
```
